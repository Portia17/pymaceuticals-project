### Analysis

- In this data, there were an almost even amount of male vs female mice making the data fairly generalizable across genders.  The drugs tested at the most timepoints were Capomulin and Ramicane, and the drugs tested at the least timepoints were Ceftamin and Propriva.  Capomulin and Ramicane appear to be more efficatious than Infubinol and Ceftamin as they had a lower average tumor size at the end but a p-value needs to be calculated to ensure these data are more than just due to chance. For mouse l509, tumor volume tended to trend down overtime.  Finally, there was a positive correlation between weight and tumor volume.


 

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single dataset
mouse_data_complete = pd.merge(mouse_metadata, study_results, on=["Mouse ID"])

# Display the data table for preview
mouse_data_complete.head()
```

```python
# Checking the number of mice.
mouse_data_complete["Mouse ID"].nunique()
```

```python
# Getting the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
has_duplicates = mouse_data_complete.duplicated(subset=["Mouse ID", "Timepoint"])

duplicates = mouse_data_complete[has_duplicates]

duplicates
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
g989 = mouse_data_complete.loc[mouse_data_complete['Mouse ID']=='g989']
g989
```

```python
mouse_data_complete["Mouse ID"].nunique()
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_mouse = mouse_data_complete.drop(g989.index)
mouse_data = cleaned_mouse.copy()
cleaned_mouse
```

```python
# Checking the number of mice in the clean DataFrame.
cleaned_mouse["Mouse ID"].nunique()
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.
mouse_mean = cleaned_mouse.groupby(["Drug Regimen"]).mean()["Tumor Volume (mm3)"]
mouse_median = cleaned_mouse.groupby(["Drug Regimen"]).median()["Tumor Volume (mm3)"]
mouse_var = cleaned_mouse.groupby(["Drug Regimen"]).var()["Tumor Volume (mm3)"]
mouse_stdv = cleaned_mouse.groupby(["Drug Regimen"]).std()["Tumor Volume (mm3)"]
mouse_sem = cleaned_mouse.groupby(["Drug Regimen"]).sem()["Tumor Volume (mm3)"]

desc_tumor_df = pd.DataFrame({
    "Mean Tumor Volume": mouse_mean,
    "Median Tumor Volume": mouse_median,
    "Tumor Volume Variance": mouse_var,
    "Tumor Volume Std. Dev.": mouse_stdv,
    "Tumor Volume Std. Err.": mouse_sem
})

desc_tumor_df
```

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, 
# and SEM of the tumor volume for each regimen

# Using the aggregation method, produce the same summary statistics in a single line.

summary_df = cleaned_mouse.groupby("Drug Regimen").agg({"Tumor Volume (mm3)":["mean", "median", "var", "std", "sem"]})
summary_df
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
last_duplicates = mouse_data.drop_duplicates(subset=["Mouse ID"], keep="last")

# Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
#merged_last= pd.merge(mouse_data, last_duplicates)

#merged_last
last_duplicates
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ["Capomulin","Ramicane","Infubinol","Ceftamin"]


# Create empty list to fill with tumor vol data (for plotting)
treatments_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
    ## Locate the rows which contain mice on each drug and get the tumor volumes
    # add subset 
    # Determine outliers using upper and lower bounds

for drug in treatments:
    tumorTreatment = last_duplicates.loc[last_duplicates['Drug Regimen']== drug, "Tumor Volume (mm3)"]
    treatments_data.append(tumorTreatment)
    quartiles = tumorTreatment.quantile([.25,.5,.75])
    lowerq = quartiles[0.25]
    upperq = quartiles[0.75]
    iqr = upperq-lowerq
    lower_bound = lowerq - (1.5*iqr)
    upper_bound = upperq + (1.5*iqr)
    #find everything below lower bound or above upper upper bound
    outlier = tumorTreatment.loc[(tumorTreatment< lower_bound) | (tumorTreatment> upper_bound)]
    print(f"{drug}'s potential outliers: {outlier}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.

red_circle = dict(markerfacecolor='red', marker='o', markersize = 10)

plt.boxplot(treatments_data, labels=treatments, flierprops=red_circle)

plt.ylabel('Final Tumor Volume (mm3)')

plt.show()
```

## Line and Scatter Plots

```python
# Generate a line plot of tumor volume vs. time point for a mouse treated with Capomulin

l509 = cleaned_mouse.loc[cleaned_mouse['Mouse ID']== 'l509']

plt.plot(l509["Timepoint"], l509["Tumor Volume (mm3)"])
plt.title("Capomulin Treatment of Mouse l509")
plt.xlabel("Timepoint (Days)")
plt.ylabel("Tumor Volume (mm3)")

plt.show()
```

```python
# Generate a scatter plot of avg tumor volume vs. avg mouse weight for the Capomulin regimen
drug_cap = cleaned_mouse.loc[cleaned_mouse['Drug Regimen']== 'Capomulin']
drug_cap2 = drug_cap.groupby("Mouse ID")[["Tumor Volume (mm3)","Weight (g)"]].mean()

drug_weight = (drug_cap2["Weight (g)"])
drug_tumor = (drug_cap2["Tumor Volume (mm3)"])

plt.scatter(drug_weight, drug_tumor)
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.show()
#drug_cap2
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and linear regression model 
# for mouse weight and average tumor volume for the Capomulin regimen
# average weight vs average tumor volume

slope, yint, r, p, ste= st.linregress(drug_weight, drug_tumor)
print(f"The correlation between mouse weight and the average tumor volume is {round(r,2)}")
reg = slope*(drug_weight)+ yint

plt.scatter(drug_weight, drug_tumor)
plt.plot(drug_weight, reg, color = "red")
plt.ylabel("Tumor Volume (mm3)")
plt.xlabel("Weight (g)")
plt.show()
```
